```python
import pandas as pd
AOS100_data = pd.read_csv("AOS100_dataset_v2.csv")
AOS100_data
```

# input（X）　output（y）の２つにデータを分ける

```python
# input set
X = AOS100_data.drop(columns=["TIME"])

X
```

```python
# output set
y = AOS100_data["TIME"]

y
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
# fitでデータをモデルに入れる
model.fit(X,y)
```

```python
# 実際に予測

predictions = model.predict([[40, 25, 60], [100, 26, 65]])
print(predictions)
```

# trainとtestにわけて予測率を出す

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)  #　４つtupleがでる

model.fit(X_train, y_train)
predictions = model.predict(X_test)


score = accuracy_score(y_test, predictions)
score
```
